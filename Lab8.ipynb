#Functie Leaky ReLu Activation

```python
def leaky_relu(x):
  if x>0 :
    return x
  else :
    return 0.01*x
x=1.0
print('Applying Leaky Relu on (%.1f) gives %.1f' % (x,leaky_relu(x)))
x=-10.0
print('Applying Leaky Relu on (%.1f) gives %.1f' % (x,leaky_relu(x)))
x=0.0
print('Applying Leaky Relu on (%.1f) gives %.1f' % (x,leaky_relu(x)))
x=15.0
print('Applying Leaky Relu on (%.1f) gives %.1f' % (x,leaky_relu(x)))
x=-20.0
print('Applying Leaky Relu on (%.1f) gives %.1f' % (x,leaky_relu(x)))
```
