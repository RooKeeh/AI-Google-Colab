Import biblioteci tensorflow si numpy


Exemplu...

```
import tensorflow as tf
import numpy as np
from tensorflow import keras
```

Selectarea modelului: o retea neuronala cu un singur neuron


```
model = tf.keras.Sequential([keras.layers.Dense(units=1, input_shape=[1])])
```

Selectarea algoritmului de optimizare si a functiei de cost

```
model.compile(optimizer='sgd', loss='mean_squared_error')
```

Incercam sa modelam datele astfel incat plecand de la valorile lui X sa obtinem estimari pentru Y

```
xs = np.array([-1.0,  0.0, 1.0, 2.0, 3.0, 4.0], dtype=float)
ys = np.array([-3.0, -1.0, 1.0, 3.0, 5.0, 7.0], dtype=float)
```

Pentru datele de mai sus, daca ar fi sa utilizam o functie determinista pentru date, am utiliza y=2x-1

Model.fit va construi modelul pe baza datelor de intrare, intr-un numar definit de pasi (epochs)

```
model.fit(xs, ys, epochs=200)
```

Realizam predictia pentru valoarea lui x=10

```
print(model.predict([20.0]))
```
